## <center>Educational Equity: Part I</center>
This is the first of a series of blog posts on educational equity.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statsmodels.api as sm
%matplotlib inline
```

```python
pov = pd.read_excel('poverty_by_district.xls')
rev = pd.read_excel('annual_survey.xls')
```

```python
pov.columns = pov.iloc[1]
pov.drop([0,1])
pov[pov.columns[5]] = pd.to_numeric(pov[pov.columns[5]],errors = "coerce")
pov[pov.columns[6]] = pd.to_numeric(pov[pov.columns[6]],errors = "coerce")

rev.NAME = rev.NAME.str.lower()
pov.Name = pov.Name.str.lower()

rev.NAME = rev.NAME.str.strip()
pov.Name = pov.Name.str.strip()
```

```python
pov['poverty_proportion'] = pov[pov.columns[6]]/pov[pov.columns[5]]

rev['rev_per_scholar'] = rev['TOTALREV']/rev['ENROLL']
```

```python
rev_pov = pd.merge(rev[['NAME','rev_per_scholar']], pov[['Name',pov.columns[7]]], left_on="NAME",right_on="Name",how = "inner")
```

```python
#Remove NA
rev_pov = rev_pov.dropna()

#Remove Outliers
rev_pov = rev_pov[rev_pov.rev_per_scholar<rev_pov.rev_per_scholar.quantile(0.95)]
rev_pov = rev_pov[rev_pov.rev_per_scholar>rev_pov.rev_per_scholar.quantile(0.05)]
```

```python
Y = rev_pov['rev_per_scholar']
X = rev_pov['poverty_proportion']
```

```python
X = sm.add_constant(X)
mod = sm.OLS(Y,X)
reg = mod.fit()

print(reg.summary())

plt.figure(1)
      
plt.plot(X.poverty_proportion,Y,'o',alpha=0.2)

plt.plot (X.poverty_proportion,reg.predict(),"-")

plt.title('Funds per Student against Poverty Rate')
plt.xlabel('Poverty Rate')
plt.ylabel('Funds per Student (thousands)')

```
