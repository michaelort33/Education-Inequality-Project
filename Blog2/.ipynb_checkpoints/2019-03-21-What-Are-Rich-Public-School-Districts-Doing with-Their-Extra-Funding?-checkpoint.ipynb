## <center>What Are Rich Public School Districts Doing with Their Extra Funding?</center>

### Intro

In the last [blog post](https://blog.depthfirsttraining.com/educational%20equity/School-Districts-with-Richer-Kids-Get-More-Public-Funding/), a simple linear regression indicated that richer school districts got more public funding for their schools. In this post, I want to breakdown school district budgets so we can see where the differences in funding across districts are felt. 

My hypothesis was that the differences in funding, while a small percentage of the overall budget, would not be evenly spread across the budget. Rather they would show up as a large percentage difference in a few key categories where the cuts are most being felt. 

To start, I am going to breakdown the budget into two parts, spending on support services and spending on instructional services. I am wondering if poorer districts still spend about the same relative amounts on support services, and as a result, have even less money to spend on instruction.

#### Imports

```python
import pandas as pd
from IPython.display import display
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from matplotlib import cm as cm
from scipy import stats
import seaborn as sns
%matplotlib inline
```

#### Reading Data

```python
pov = pd.read_excel('../Data/poverty_by_district.xls')
an_su = pd.read_excel('../Data/annual_survey.xls')
descriptions = pd.read_excel('../Data/annual_survey_index.xls')

display(an_su.head(3))
display(pov.head(3))
```

```python
#The descriptions of column headers in an_su are listed in descriptions
display(descriptions.tail(11))
```

#### Feature Creation
School spending is broken down pretty well so there are a lot of interesting variables here but just to start, we'll look at PPSTOTAL and PPITOTAL, spending on support and spending on instruction.

```python
#Prep for merge, make key for merge lowercase for both data sets
an_su.NAME = an_su.NAME.str.lower()
pov.Name = pov.Name.str.lower()

an_su.NAME = an_su.NAME.str.strip()
pov.Name = pov.Name.str.strip()

#Create poverty proportion feature
pov['poverty_proportion'] = pov[pov.columns[6]] / pov[pov.columns[5]]


an_pov = pd.merge(an_su[['NAME','PPITOTAL','PPSTOTAL']], pov[['Name',pov.columns[7]]],
                   left_on = "NAME",right_on = "Name",how = "inner")
```

#### Data Cleaning

```python
#Remove NA
an_pov = an_pov.dropna()

#There shouldn't be zero values in either variable
(an_pov[['PPSTOTAL','PPITOTAL']] == 0).sum()
```

#### Distributions
I'm going to check out the distribution of our features.

```python
#Plot a histogram
plt.figure(5,figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
bins = 200
plt.hist(an_pov.PPSTOTAL.values,bins, label = 'PPSTOTAL')
plt.hist(an_pov.PPITOTAL.values,bins, label = 'PPIOTAL',alpha=0.5)
plt.legend(loc = 'upper right')
plt.ylabel('Frequency')
plt.xlabel('Spending')

plt.title('Spending Histogram')
plt.xlabel('Spending')
plt.ylabel('Frequency')

#boxplots
fig, ax = plt.subplots(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
ax.boxplot(an_pov[['PPSTOTAL','PPITOTAL']].values)
ax.set(xticklabels = ['PPS','PPI'])

plt.title('Boxplots of Spending on Support and Instruction')
plt.xlabel('Spending')
plt.ylabel('Funds per Student')

Q1 = an_pov.quantile(0.25)
Q3 = an_pov.quantile(0.75)
IQR = Q3 - Q1
```

#### Outliers

The distributions reveal very right skewed variables with some nonsensical outliers. I'm going to plot some of these outliers, points outside of $1.5 \times IQR$.

```python
#Save the outliers and plot them
outliers = an_pov[((an_pov[['PPSTOTAL','PPITOTAL']] < (Q1 - 1.5 * IQR)) | (an_pov[['PPSTOTAL','PPITOTAL']] > (Q3 + 1.5 * IQR))).any(axis=1)]

plt.figure(0,figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(outliers.poverty_proportion.values,outliers.PPSTOTAL.values,'o',alpha=0.2)
plt.title('Support Spending against Poverty Rate')
plt.xlabel('Poverty Rate')
plt.ylabel('Frequency')

plt.figure(1,figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(outliers.poverty_proportion.values,outliers.PPITOTAL.values,'o',alpha=0.2)

plt.title('Instructional Spending against Poverty Rate')
plt.xlabel('Poverty Rate')
plt.ylabel('Frequency')
```

#### Removing Outliers

I'm going to remove these outliers. I wouldn't normally just throw away outliers, but the data set here shouldn't vary too much, and I would guess that most of these outliers are reporting errors. For example, there shouldn't be any school with spending per student at \$100,000 or 0. This seems to be confirmed in the plots of the outliers which show them evenly distributed across districts. 

```python
#Remove outliers of 1.5 IQR
an_pov = an_pov[~((an_pov[['PPSTOTAL','PPITOTAL']] < (Q1 - 1.5 * IQR)) | (an_pov[['PPSTOTAL','PPITOTAL']] > (Q3 + 1.5 * IQR))).any(axis=1)]

#Remove the rows with zero
an_pov = an_pov[an_pov.PPSTOTAL != 0]
an_pov = an_pov[an_pov.PPITOTAL != 0]

#Remove values >3 sd away
z = np.abs(stats.zscore(an_pov.PPSTOTAL.values))
an_pov = an_pov[z < 3]

z = np.abs(stats.zscore(an_pov.PPITOTAL.values))
an_pov = an_pov[z < 3]
```

#### New Distributions

Without the outliers, the histograms and boxplots look much nicer. 

```python
#Histogram
plt.figure(2,figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
bins = 200
plt.hist(an_pov.PPSTOTAL.values,bins, label = 'PPSTOTAL')
plt.hist(an_pov.PPITOTAL.values,bins, label = 'PPITOTAL',alpha=0.5)
plt.legend(loc = 'upper right')
plt.ylabel('Frequency')
plt.xlabel('Spending')

plt.title('Spending Histogram')
plt.xlabel('Spending')
plt.ylabel('Frequency')

#boxplots
fig, ax = plt.subplots(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
ax.boxplot(an_pov[['PPSTOTAL','PPITOTAL']].values)
ax.set(xticklabels = ['PPS','PPI'])

plt.title('Boxplots of Spending on Support and Instruction')
plt.xlabel('Spending')
plt.ylabel('Funds per Student')
```

#### Correlations

Before the regressions let's look at the correlation matrix of the three variables. I was expecting large, negative correlations. Sadly, I was wrong. 

```python
corr = an_pov[['PPSTOTAL','PPITOTAL','poverty_proportion']].corr()
fig, ax1 = plt.subplots(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
cmap = cm.get_cmap('jet', 30)
cax = ax1.imshow(corr, interpolation="nearest", cmap=cmap)
ax1.grid(True)
plt.title('School Spending Correlations')
plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns);
fig.colorbar(cax, ticks=np.linspace(0,1,10).round(2))
```

#### Simple Linear Regressions

Running the actual SLR models, we see a very small positive relation between poverty rate and support services.

```python
#Regression model for PPS
Y1 = an_pov['PPSTOTAL'].astype(int)
X = an_pov['poverty_proportion']

X = sm.add_constant(X)
mod1 = sm.OLS(Y1,X)
reg1 = mod1.fit()

print(reg1.summary())

#Plot of regression1
plt.figure(2,figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')

plt.plot(X.poverty_proportion,Y1,'o',alpha=0.2)

plt.plot (X.poverty_proportion,reg1.predict())

plt.title('PPSTOTAL against Poverty Rate')
plt.xlabel('Poverty Rate')
plt.ylabel('Funds per Student')

```

There is a moderate negative relation between poverty rate and instructional spending:

```python
#Regression model for PPI
Y2 = an_pov['PPITOTAL'].astype(int)

mod2 = sm.OLS(Y2,X)
reg2 = mod2.fit()
print(reg2.summary())

#Plot of regression2

plt.figure(3,figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')

plt.plot(X.poverty_proportion,Y2,'o',alpha=0.2)

plt.plot (X.poverty_proportion,reg2.predict())

plt.title('PPITOTAL against Poverty Rate')
plt.xlabel('Poverty Rate')
plt.ylabel('Funds per Student')
```

### Conclusion

So overall these results are kind of what I expected but they are also somewhat surprising. Even though, overall, poorer schools get less funding, they have to spend more (or about the same) on support services. This makes sense because there are additional costs associated with poverty. ([See this amazing analysis](http://apps.urban.org/features/school-funding-do-poor-kids-get-fair-share/).) Therefore, these schools should be at an even larger disadvantage in other areas than the overall funding differences seem to imply. This is where the results get strange. Even though we did find that spending on instructional services was lower in higher poverty areas, it had about the same slope as the regression on overall spending. Given that spending on support services was about the same, I would expect the deficit in spending on instructional services to be even larger than the overall deficits imply. I guess we are missing key areas of the budget. Stay tuned!
