### Imports

```python
from keras.layers import Dense, LSTM, Dropout
from keras.callbacks import EarlyStopping
from keras.models import load_model, Sequential
from keras.constraints import max_norm

import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from matplotlib import gridspec
import pandas as pd

# Autocomplete
%config IPCompleter.greedy=True
```

### Reading and setting up the data

```python
# reading the data with pandas
data = pd.read_csv('../input/march_april_btc_minute.csv')

# divide data with keep, formate DateTime and set Date to index
def read(df, keep=1):
    # Use subset of data (my little gpu can't do much)
    cut = round(len(df) / keep)
    df = df.iloc[:cut, :]

    # convert to date type
    df.Date = df.Date.astype('datetime64[ns]')

    # Use Date column as index
    df = df.set_index('Date', drop=True)
    
    return df

# read the data and only use the first half of it (my gpu is slow)
data = read(data, keep = 20)

# Plot volume and price on two y axes
# Takes the column names of x, y1, y2, and a DF
def plot_2_y(x, y1, y2, df):
    fig, host = plt.subplots()
    fig.subplots_adjust(right=0.75)
    plt.xticks(rotation=60)
    par1 = host.twinx()

    p1, = host.plot(df[y1], "k-", label=y1)
    p2, = par1.plot(df[y2], "b-", label=y2, alpha=0.5)

    host.set_xlabel(x)
    host.set_ylabel(y1)
    par1.set_ylabel(y2)
    lines = [p1, p2]
    host.legend(lines, [l.get_label() for l in lines])

    plt.show()


# Plot the data
plot_2_y('Date', 'price', 'volume', data)
```

### Create features and clean data

```python
# Normally you will probably want to add rolling means
# that go back farther than 60 minutes.
def create_features(df):
    # take the last three minutes as inputs for volume and price
    for i in list(range(1, 2)):
        df['price_shift' + str(i)] = df.price.shift(i)

    return df


# Create the features
data = create_features(data)

# This function is really just drop na but I'm thinking there
# are probably other things I will want to add to clean data soon. 
def clean(df):
    # Remove NAs hat were introduced with the shift
    # in create features
    df = df.dropna()
    
    # drop unnecessary columns
    df = df.drop(['volume'], axis=1)

    return df


data = clean(data)
```

### Splitting train and test, plotting data

```python
# 80 20 split for train and test
def split_train_test(df, split=0.8):
    index = round(split * len(df))
    df_train = df.iloc[:index, :]
    df_test = df.iloc[index:, :]

    return df_train, df_test


# split the test and train
train, test = split_train_test(data)

plt.figure(1)
plt.subplot(111)
plt.xticks(rotation=60)
plt.plot(train.price)
plt.plot(test.price)
plt.legend(['Train', 'Test'])
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

```

# Scaling data

Scaling data is important for training LSTMs because having very different scales on inputs can cause training to become very slow and also increases the likelihood of getting stuck at a local minimum. 

However, scaling this sort of time series data is annoying because we can expect the training data to have a very different range than the test data. Bitcoin often increases or decreases drastically. Even if we trained on Bitcoin's entire history, we can likely get a test set in the future that goes outside of its historical range. 

To solve this scaling issue I can think of two things to do:

1) Switch our inputs from actual prices to change in price from previous price, and predict change in price instead of the actual price. (You can just convert this back to actual price after.) This change input would be limited to at most [0,1], no matter what the future holds in store. 

2) Standardize input data instead of scaling it. If you can estimate a mean and standard deviation of test data then you can create a model that is more robust to predictions of future data that is outside the range of the training data. I tried this and it didn't work. Maybe I did something but my results were terrible because the test set is at a random massive leap well outside of the range of the training data. 

```python
# Normalize data
scaler = MinMaxScaler()
scaler.fit(train)
train = scaler.transform(train)
test_sc = scaler.transform(test)
```

### Train model

```python
def prep_neural_data_format(df):

    # split train to x and y
    train_x = df[:, 1:]
    train_y = df[:, 0]

    # reshape to 3d for lstm input
    train_x = train_x.reshape(train_x.shape[0], train_x.shape[1], 1)

    return train_x, train_y

neural_x, neural_y = prep_neural_data_format(train)
```

```python
def train_model(train_x, train_y):

    # Create LSTM model
    batch_size = 2            
    epochs = 15               
    seq_len = 1             
    loss='mean_squared_error' 
    optimizer = 'rmsprop'     
    activation = 'linear'     
    input_shape=(train_x.shape[1],1)      
    output_dim = 1           
    model = Sequential()
    model.add(LSTM(units=output_dim, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=output_dim, return_sequences=True))
    model.add(LSTM(units=output_dim, return_sequences=True))
    model.add(LSTM(units=output_dim, return_sequences=False))
    model.add(Dense(units=1,activation=activation))
    model.compile(optimizer=optimizer,loss=loss)
    early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)
    model.fit(x=train_x,
          y=train_y,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=0.05,
          callbacks=[early_stop])
    
    # Save the trained model to file
    # model.save('btc_predictor.h5') # Uncomment to save the model
    
    return model
    

# train the model
model = train_model(neural_x, neural_y)

# Load trained model from file
# model = load_model('btc_predictor_5.h5') # Uncomment this if you saved the model above
```

```python
def prep_test_data(df):

    # Select x matrix
    df_x = df[:, 1:]
    df_y = df[:, 0]

    # Reshape for prediction
    df_x = df_x.reshape(df_x.shape[0], df_x.shape[1], 1)

    return df_x, df_y


test_x_sc, test_y_sc = prep_test_data(test_sc)

# Measure error
lstm_test_mse = model.evaluate(test_x_sc, test_y_sc)
print('LSTM: %f' % lstm_test_mse)

# predictions
predictions_sc = model.predict(test_x_sc)

# Reshape test_x_c back to 2D
test_x_sc = test_x_sc.reshape(test_x_sc.shape[0], test_x_sc.shape[1])

#Combine predictions with test_x_sc to inverse transform

prediction_matrix_sc = np.hstack((predictions_sc,test_x_sc))

prediction_matrix = scaler.inverse_transform(prediction_matrix_sc)

predictions = prediction_matrix[:,0]

# plot predictions and actual values
plt.cla()
plt.figure(2)
plt.plot(predictions)
plt.plot(test.price.values, alpha = 0.7)
plt.legend(['predictions', 'actual'])

# Print the MSE of the LSTM model
print((np.mean((predictions - test.price.values)**2)))

# Now compare this model to a baseline of a rolling average of 3 previous minutes
def compare_rolling_mean(data):
    data['price_rolling'] = data.price.rolling(window=30).mean().shift()
    data = data.dropna()
    train, test = split_train_test(data)
    plt.figure(3)
    plt.plot(test.price)
    plt.plot(test.price_rolling)
    print(np.mean(((test.price_rolling.values - test.price.values)**2)))
    
compare_rolling_mean(data)
```
